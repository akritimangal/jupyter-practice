# is keyword

```python
a = 1000
b = 1000

a == b
a is b
'''Note

== → compares values

is → compares memory identity

Large integers create separate objects
correct use
if x is None:
    ...
'''
```

```python
a = 5
b = 5
a is b
'''Note

Python caches small integers (-5 to 256)

Same object reused for performance

This behavior is an implementation detail, not logic'''
```

# Boolean as Integers


```python
a = False
b = 0
type(a + b)
'''Note

False behaves like 0

True behaves like 1

Boolean arithmetic results in int'''
```

```python
x = 10
print(type(type(x)))
'''

Note

type(x) returns a class

Classes themselves are objects of type type'''
```
