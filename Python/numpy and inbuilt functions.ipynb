# Slices are views, not copies

```python
a = np.array([1,2,3,4])
b = a[1:3]
b[0] = 99
print(a)
```

# To force copy:

```python
ar=np.array([1,3,4,7])
br = ar[1:3].copy()
br[0] = 99
print(ar)
```
